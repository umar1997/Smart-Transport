### Shortest Path Algorithm

```python
# https://stackoverflow.com/questions/43442919/floyd-warshall-algorithm-get-the-shortest-paths
# https://docs.scipy.org/doc/scipy/reference/generated/scipy.sparse.csgraph.floyd_warshall.html#scipy.sparse.csgraph.floyd_warshall
from scipy.sparse import csr_matrix
from scipy.sparse.csgraph import floyd_warshall

def ConstructPath(p, i, j):
    path = []
    while i!=j:
        path.append(Index2Loc[j])
        j = p[i,j]
    path.append(Index2Loc[j])
    return path[::-1]
```

### Supply Parameters

```python
def time2timedelta(timeObj):
    return datetime.datetime.combine(datetime.date.min, timeObj) - datetime.datetime.min

def timedelta2time(deltaObj):
    return (datetime.datetime.min + deltaObj).time()

def datetime2time(timeObj):
    return datetime.datetime.strptime(timeObj.strftime('%H:%M:%S'),'%H:%M:%S').time()
```

```python
Bus_Capacity = 10
Fleet_Size = 10
Uniform_Speed = 50
Waiting_Time_Threshold = time2timedelta(datetime.time(0,12)) # 12 minutes
Travel_Time_Threshold = time2timedelta(datetime.time(0,18)) # 24 minutes
```

```python
import math

def toTime(random_number):
    hours = math.floor(random_number/60)
    minutes = math.floor(random_number) if random_number < 60 else (math.floor(random_number) - (hours*60))
    seconds = round(((random_number - math.floor(random_number))*0.6)*100)
    x = timedelta2time(datetime.timedelta(hours = hours, minutes = minutes, seconds = seconds))
    return x 

def timeCalculation(Uniform_Speed, Distance):
    return (Distance/Uniform_Speed)*60

toTime(timeCalculation(Uniform_Speed, 5))
```

### Creating Travel Demand

```python
def random_date(start, end):
    """
    This function will return a random datetime between two datetime 
    objects.
    """
    delta = end - start
    int_delta = (delta.days * 24 * 60 * 60) + delta.seconds
    random_second = randrange(int_delta)
    return start + datetime.timedelta(seconds=random_second)
```

```python
d1 = datetime.datetime.strptime('1/1/2009 9:00 AM', '%m/%d/%Y %I:%M %p')
d2 = datetime.datetime.strptime('1/1/2009 9:30 PM', '%m/%d/%Y %I:%M %p')


N = 200
arrival_times = []
passenger_origin = []
passenger_dest = []

for i in range(N):
    arrival_times.append(random_date(d1, d2))
    passenger_origin.append(generateOrigin())
    passenger_dest.append(generateDestination(passenger_origin[-1]))
    
TD = pd.DataFrame()
TD['Passengers'] = list(np.arange(N) + 1)
TD['Entry Time'] = arrival_times
TD['Entry Time'] = TD['Entry Time'].sort_values(ascending=True).tolist()
TD['Passenger Origin'] = passenger_origin
TD['Passenger Destination'] = passenger_dest

TD.head(50)
```

```python
from heapq import nsmallest
def second_smallest(numbers):
    return nsmallest(2, numbers)[-1]

from collections import OrderedDict 
def sortKeys(d):
    d = OrderedDict(sorted(d.items())) 
    return {**d}

def sortValues(d):
    d = OrderedDict(sorted(d.items(), key=lambda x: x[1]))
    return {**d}

import copy
import json

def get_key(val, my_dict):
    list_of_keys = []
    boolValue = 0
    for key, value in my_dict.items(): 
         if val == value: 
                list_of_keys.append(key)
    if len(list_of_keys) == 0:
        return []
    else:
        return list_of_keys
```

```python
passenger, Origin_List, Dest_List, EntryTime, waitingTime, BoardingTime, TravelTime, AlightingTime, Bus_Selected, Buses_Route_Person, Buses_Route_Loc = [], [], [], [], [], [], [], [], [], [], []
Entry, Origin, Dest = 0, 0, 0
for i in range(len(TD)):

    LastEntry = Entry
    LastDest = Dest
    LastOrigin = Origin
    
    Origin = TD['Passenger Origin'][i]
    Dest = TD['Passenger Destination'][i]
    Entry = TD['Entry Time'][i].to_pydatetime()
    

    #******************************************************************************************************************
    # Checking if any of the times in the Alight are more than the time recently realized then we will have to 
    # remove all those passengers whose alight time was in between the last demand realized time and this time realized
    # and if the alight time is more than the current realized time the current location of the bus will be the last place 
    # a person was alighted
    
    print('################################### ', i)
    print(i, LastEntry, LastOrigin, LastDest)
    display(Buses)
    
    for b in range(Fleet_Size):
        if len(Buses[b]['Dest']) == 0:
            continue
        else:
            assert len(Buses[b]['Dest'].keys()) == len(Buses[b]['Entry'].keys()) == len(Buses[b]['Alight'].keys()) == len(Buses[b]['Board'].keys())
            
            Buses[b]['Alight'] = sortValues(Buses[b]['Alight'])
            
            busAlight = copy.deepcopy(Buses[b]['Alight'])
            busBoard = copy.deepcopy(Buses[b]['Board'])
            busOrigin = Buses[b]['Origin']
            
            last_index = 0
            for ii, t in enumerate(Buses[b]['Routes']['Time']):
                if Entry >= t:
                    last_index = ii
                else:
                    break
            Buses[b]['current'] = Buses[b]['Routes']['Loc'][last_index]
                
            
            for person , time in busBoard.items():
                
                if Entry >= time and Entry >= Buses[b]['Alight'][person]:
                    passenger.append(person)
                    Bus_Selected.append(b)
                    Origin_List.append(Buses[b]['Origin'][person])
                    Dest_List.append(Buses[b]['Dest'][person])
                    
                    EntryTime.append(Buses[b]['Entry'][person])
                    BoardingTime.append(Buses[b]['Board'][person])
                    AlightingTime.append(Buses[b]['Alight'][person])
                    
    
                    waitingTime.append(timedelta2time(BoardingTime[-1] - EntryTime[-1]))
                    TravelTime.append(timedelta2time(AlightingTime[-1] - BoardingTime[-1]))
                    
                    Buses_Route_Person.append(Buses[b]['Routes']['Person'])
                    Buses_Route_Loc.append(Buses[b]['Routes']['Loc'])
#                     Buses[b]['current'] = Buses[b]['Dest'][person]
                    
                    try:
                        Buses[b]['Entry'].pop(person)
                        Buses[b]['Board'].pop(person)
                        Buses[b]['Alight'].pop(person)
                        Buses[b]['Dest'].pop(person)
                        Buses[b]['Origin'].pop(person)
                        Buses[b]['Promise_Alight'].pop(person)
                        Buses[b]['Promise_Board'].pop(person)

                    except KeyError:
                        print('Key Not In The Dictionary')
                    
                elif Entry >= time:
                    continue
#                     Buses[b]['current'] = Buses[b]['Origin'][person]
                else:
                    break
                    
            Buses[b]['Alight'] = sortValues(Buses[b]['Alight'])
    
    #******************************************************************************************************************
    # Here we look at the bus selected for the current demand realized, we first check the cost of without including the person
    # and then we check when we add the person what is the difference between the two costs. The minimum of the two costs bus is
    # selected to pick up that passenger
    
    cost_old, cost_new, cost_diff, PathDistances, CapacityFlag = [], [], [], [], []

    for b in range(Fleet_Size):
        if len(Buses[b]['Dest']) > Bus_Capacity: # If a bus has reached full capacity
            #-------------------------------------------------------------------------------------------------------------
            
            cost_old.append(datetime.timedelta(seconds=0))
            cost_new.append(datetime.timedelta(seconds=0))
            cost_diff.append(datetime.timedelta(seconds=99999999))
            
            path_distances = {'Loc': [], 'Dist': []}
            PathDistances.append(path_distances)
            
            CapacityFlag.append(1)
            
        elif len(Buses[b]['Dest']) == 0: # If a bus is empty
            #-------------------------------------------------------------------------------------------------------------

            Bus_Origin = Buses[b]['current']
            
            
            if Origin == Bus_Origin:
                path_origin = []
                distance_dest = dist_matrix[Loc2Index[Origin]][Loc2Index[Dest]]
                path_dest = ConstructPath(predecessors, Loc2Index[Origin], Loc2Index[Dest])
            else:
                distance_origin = dist_matrix[Loc2Index[Bus_Origin]][Loc2Index[Origin]]
                path_origin = ConstructPath(predecessors, Loc2Index[Bus_Origin], Loc2Index[Origin])
                distance_dest = distance_origin + dist_matrix[Loc2Index[Origin]][Loc2Index[Dest]]
                path_dest = path_origin + ConstructPath(predecessors, Loc2Index[Origin], Loc2Index[Dest])[1:]
            
            
            path_distances = {'Person': [], 'Loc': [], 'Dist': []}
            path_distances['Person'] = [-1] * len(path_dest)
            if len(path_origin) == 0:
                path_distances['Person'][0], path_distances['Person'][-1] = TD['Passengers'][i], TD['Passengers'][i] 
            else:
                path_distances['Person'][len(path_origin)-1], path_distances['Person'][-1] = TD['Passengers'][i], TD['Passengers'][i] 
            path_distances['Loc'] += path_dest
            Temp_Origin = Bus_Origin
            mini_dist = 0
            for Location in  path_distances['Loc']:
                mini_dist = mini_dist + dist_matrix[Loc2Index[Temp_Origin]][Loc2Index[Location]]
                path_distances['Dist'] += [mini_dist]
                Temp_Origin = Location
            
            assert len(path_distances['Dist']) == len(path_distances['Loc']) == len(path_distances['Person'])
            print('PD: ', b, path_distances)
            PathDistances.append(path_distances)
            
            
            cost_old.append(datetime.timedelta(seconds=0))
            cost_new.append(time2timedelta(toTime(timeCalculation(Uniform_Speed, distance_dest))))
            cost_diff.append(cost_new[-1] - cost_old[-1])
            
        else: # If a bus is NOT empty
            #-------------------------------------------------------------------------------------------------------------
            
            assert len(Buses[b]['Dest'].keys()) == len(Buses[b]['Entry'].keys()) == len(Buses[b]['Alight'].keys())
            
            passengers = list(Buses[b]['Dest'].keys())
            sum = datetime.timedelta(seconds=0)
            for p in passengers:
                entryTime = datetime2time(Buses[b]['Entry'][p])
                alightTime = datetime2time(Buses[b]['Alight'][p])
                sum = sum + (time2timedelta(alightTime) - time2timedelta(entryTime))

            cost_old.append(sum)
            
            ##########################################################################
            
            Bus_Origin = Buses[b]['current']
            
            # Getting 3 columns of the passenger, the origin and the destination including the latest passenger
            path_distances = {'Person': [], 'Loc': [], 'Dist': []}
            fromLoc, toLoc, person = [], [], []
            set_origins = set(list( Buses[b]['Origin'].values()))
            for v in set_origins:
                ppl = get_key(v, Buses[b]['Origin'])
                if len(ppl):
                    for p in ppl:
                        person.append(p)
                        fromLoc.append(v)
                        toLoc.append(Buses[b]['Dest'][p])
            person.append(TD['Passengers'][i])
            fromLoc.append(Origin)
            toLoc.append(Dest)
            
            # Adding all the people staring from the Bus Origin into our path distances
            # If multiple are starting at the bus origin
            flags = [0] * len(toLoc)
            indices_origin = [index for index, value in enumerate(fromLoc) if value == Bus_Origin]
            ppl_index_boarding = []
            if len(indices_origin) != 0:
                for i_o in indices_origin:
                    flags[i_o] = 1
                    ppl_index_boarding.append(i_o)
                    path_distances['Person'].append(person[i_o])
                    path_distances['Loc'].append(Bus_Origin)
                    path_distances['Dist'].append(0.0)
                # Below we dont add flags cause no one is a part of toLoc
                assert len(toLoc) == len(fromLoc) == len(person) == len(flags)
            else:
                path_distances['Person'].append(-1)
                path_distances['Loc'].append(Bus_Origin)
                path_distances['Dist'].append(0.0)
                assert len(toLoc) == len(fromLoc) == len(person)
            
            origin_boarded = len(ppl_index_boarding)
            
            print('List',toLoc, fromLoc, person, flags)
            
            total_dist = 0
            Temp_Origin = Bus_Origin
            check_points = set(fromLoc)
            # If someones Origin is the Bus Origin then add their destination as well and remove him from the check points
            ppl = get_key(Bus_Origin, Buses[b]['Origin'])
            if Origin == Bus_Origin:
                ppl += [TD['Passengers'][i]]
                if len(ppl) > 1:
                    for p in ppl:
                        if p == TD['Passengers'][i]:
                            check_points.add(Dest)
                        else:
                            check_points.add(Buses[b]['Dest'][p])
                    check_points.remove(Bus_Origin)
                else:
                    check_points.add(Dest)
                    check_points.remove(Bus_Origin)
            else:
                if len(ppl):
                    for p in ppl:
                        check_points.add(Buses[b]['Dest'][p])
                    check_points.remove(Bus_Origin)
            
            check_points = list(check_points)
            
            # Find the nearest neighbour of all the origins and the destinations whose origins have been visted
#---------------  
            ppl_comp_boarding, ppl_comp_leaving = [], []
            indices_start = [index for index, value in enumerate(fromLoc) if value == Temp_Origin]
            for i_s in indices_start:
                ppl_comp_boarding.append(person[i_s])
            while len(check_points)!= 0:
                smallest_checkpoint_dist = []
                mini_path_bool = 0
                for cp in check_points:
                    smallest_checkpoint_dist.append(dist_matrix[Loc2Index[Temp_Origin]][Loc2Index[cp]])
                    if Temp_Origin == cp:
                        mini_path_bool = 1
                smallest_index = smallest_checkpoint_dist.index(min(smallest_checkpoint_dist)) 
                if mini_path_bool:
                    mini_path = [Temp_Origin]
                else:
                    mini_path = ConstructPath(predecessors, Loc2Index[Temp_Origin], Loc2Index[check_points[smallest_index]])[1:]
                Temp_Origin = check_points[smallest_index]
                
                indices_origin = [index for index, value in enumerate(fromLoc) if value == Temp_Origin]
                add_dest = [value for index, value in enumerate(toLoc) if index in indices_origin]
                
                # Mark flag of those who have boarded and add the passengers boarded in a list
                ppl_index_boarding = []
                if len(indices_origin) != 0:
                    for i_o in indices_origin:
                        if person[i_o] not in ppl_comp_boarding:
                            flags[i_o] = 1
                            ppl_index_boarding.append(person[i_o])
                        
                for d in add_dest:
                    if d not in check_points:
                        check_points.append(d)
                
                # delete the people whose has been achieved and add all the complete passengers in a list
                ppl_index_leaving = []
                capture_id = []
                if Temp_Origin in toLoc:
                    indices_dest = [index for index, value in enumerate(toLoc) if value == Temp_Origin]
                    
                    for i_d in indices_dest:
                        if flags[i_d] == 1:
                            if person[i_d] not in ppl_comp_leaving:
                                ppl_index_leaving.append(person[i_d])
                                capture_id.append(i_d)
                    
                    toLoc = [value for index, value in enumerate(toLoc) if index not in capture_id]
                    fromLoc =[value for index, value in enumerate(fromLoc) if index not in capture_id]
                    person = [value for index, value in enumerate(person) if index not in capture_id]
                    flags = [value for index, value in enumerate(flags) if index not in capture_id]
                     
                    assert len(toLoc) == len(fromLoc) == len(person) == len(flags)
                 
                assert mini_path[-1] == Temp_Origin  
                   
                ppl_comp_boarding += ppl_index_boarding
                ppl_comp_leaving += ppl_index_leaving
                    
                # Add those many people to the end who had boarded and who had gotten out and make sure Person, Dist and Path are all the same
                mini_path = mini_path[:-1]
                person_list = [-1] * len(mini_path)
                for p in ppl_index_leaving:
                    person_list.append(p)
                    mini_path.append(Temp_Origin)
                for p in ppl_index_boarding:
                    person_list.append(p)
                    mini_path.append(Temp_Origin)
                
                path_distances['Person'] += person_list
                path_distances['Loc'] += mini_path
                check_points.pop(smallest_index)
#---------------
         
            # From the Loc in path distances add the dist from the dist_matrix 
            mini_dist = 0
            Temp_Origin = Bus_Origin
            if path_distances['Person'][0] == -1:
                for Location in  path_distances['Loc'][(origin_boarded+1):]:
                    mini_dist = mini_dist + dist_matrix[Loc2Index[path_distances['Loc'][0]]][Loc2Index[Location]]
                    path_distances['Dist'] += [mini_dist]
                    Temp_Origin = Location
            else:
                for Location in  path_distances['Loc'][origin_boarded:]:
                    mini_dist = mini_dist + dist_matrix[Loc2Index[Temp_Origin]][Loc2Index[Location]]
                    path_distances['Dist'] += [mini_dist]
                    Temp_Origin = Location
            
            assert len(path_distances['Dist']) == len(path_distances['Loc']) == len(path_distances['Person'])
            print('PD: ',b, path_distances, '-------')
            PathDistances.append(path_distances)
            
            ##########################################################################
            
            # Find the cost of adding the new passenger when adding him to the list of current passengers
            wontPick = 0
            sum = datetime.timedelta(seconds=0)
            for p in passengers:
                entryTime = Buses[b]['Entry'][p]
                
                enter = Buses[b]['Origin'][p]
                exit = Buses[b]['Dest'][p]
                
                # get the index for the passenger as to where he got on and where he got off
                enter_index, exit_index = 0, 0
                count_occurence = 0
                for ii, pp in enumerate(path_distances['Person']):
                    if pp == p and count_occurence == 1:
                        count_occurence += 1
                        exit_index = ii
                    elif pp == p:
                        count_occurence += 1
                        enter_index = ii
                    else:
                        continue
                
                assert count_occurence == 2
                
                assert path_distances['Loc'][enter_index] == enter
                assert path_distances['Loc'][exit_index] == exit
                
                boardTime = Entry + time2timedelta(toTime(timeCalculation(Uniform_Speed, path_distances['Dist'][enter_index])))
                alightTime = Entry + time2timedelta(toTime(timeCalculation(Uniform_Speed, path_distances['Dist'][exit_index])))
                
                # Make sure the travel time and waiting time of each person doesnt cross the limit
                if boardTime > Buses[b]['Promise_Board'][p]:
                    wontPick = 1
                 
                if alightTime > Buses[b]['Promise_Alight'][p]:
                    wontPick = 1
                    
                sum = sum + (time2timedelta(datetime2time(alightTime)) - time2timedelta(datetime2time(Buses[b]['Entry'][p])))
            
            count_occurence = 0
            for ii, pp in enumerate(path_distances['Person']):
                if pp == TD['Passengers'][i] and count_occurence == 1:
                    count_occurence += 1
                    exit_index = ii
                elif pp == TD['Passengers'][i]:
                    count_occurence += 1
                    enter_index = ii
                else:
                    continue

            assert count_occurence == 2

            
            alightTime = Entry + time2timedelta(toTime(timeCalculation(Uniform_Speed, path_distances['Dist'][exit_index])))
            sum = sum + (time2timedelta(datetime2time(alightTime)) - time2timedelta(datetime2time(Entry)))
            cost_new.append(sum)
            
            if wontPick:
                cost_diff.append(datetime.timedelta(seconds=99999999))
            else:
                cost_diff.append(cost_new[-1] - cost_old[-1])
        
    #******************************************************************************************************************
    # If the travelling time for any customer is exceeded because of addition of new passenger and this be the case for all Buses

    assert len(cost_diff) == len(cost_old) == len(cost_new)
    
    if min(cost_diff) == datetime.timedelta(seconds=99999999):
        print('Customer {} has been rejected'.format(i))
    bus_index = cost_diff.index(min(cost_diff))
    path_distances = PathDistances[bus_index]

    
    # Here we will update the bus that has been selected to pick up the passenger, that means addition of the customer
    # and adding his entry time, calculating his boarding time and all the alighting time after the route change
    print('PATH DISTANCE SELECTED: ',bus_index ,path_distances)
    if [1]* Fleet_Size != CapacityFlag: # If all Buses haven't reached full capacity
    
        Buses[bus_index]['Dest'][TD['Passengers'][i]] = Dest
        Buses[bus_index]['Entry'][TD['Passengers'][i]] = Entry
        Buses[bus_index]['Origin'][TD['Passengers'][i]] = Origin
        
        Bus_Origin = Buses[bus_index]['current']
        
        
        for ppl in list(Buses[bus_index]['Origin'].keys()):
            loc = Buses[bus_index]['Origin'][ppl]
            enter_index, exit_index, count_occurence = 0, 0, 0
            for ii, pp in enumerate(path_distances['Person']):
                if pp == ppl and count_occurence == 1:
                    count_occurence += 1
                    exit_index = ii
                elif pp == ppl:
                    count_occurence += 1
                    enter_index = ii
                else:
                    continue
            
            assert count_occurence == 2
            
            Buses[bus_index]['Board'][ppl] = Entry + time2timedelta(toTime(timeCalculation(Uniform_Speed, path_distances['Dist'][enter_index])))
            if ppl == TD['Passengers'][i]:
                Buses[bus_index]['Promise_Board'][TD['Passengers'][i]] = Buses[bus_index]['Board'][TD['Passengers'][i]] + Waiting_Time_Threshold
        
        for ppl in list(Buses[bus_index]['Dest'].keys()):
            loc = Buses[bus_index]['Dest'][ppl]
            enter_index, exit_index, count_occurence = 0, 0, 0
            for ii, pp in enumerate(path_distances['Person']):
                if pp == ppl and count_occurence == 1:
                    count_occurence += 1
                    exit_index = ii
                elif pp == ppl:
                    count_occurence += 1
                    enter_index = ii
                else:
                    continue
            
            assert count_occurence == 2
            
            Buses[bus_index]['Alight'][ppl] = Entry + time2timedelta(toTime(timeCalculation(Uniform_Speed, path_distances['Dist'][exit_index])))
            if ppl == TD['Passengers'][i]:
                Buses[bus_index]['Promise_Alight'][TD['Passengers'][i]] = Buses[bus_index]['Alight'][TD['Passengers'][i]] + Travel_Time_Threshold

        Buses[bus_index]['Routes']['Person'] = path_distances['Person']
        Buses[bus_index]['Routes']['Loc'] = path_distances['Loc']
        Buses[bus_index]['Routes']['Time'] = [Entry + time2timedelta(toTime(timeCalculation(Uniform_Speed, d))) for d in path_distances['Dist']]
        
        Buses[bus_index]['Alight'] = sortValues(Buses[bus_index]['Alight'])
    else:
        print('All Buses have reached full capacity.')



        
        
        
        
        

# Here we are dropping of the rest of the passengers that are left after all the demand that has to come has come
# Now there is no more need to pick up anymore passengers
        
for b in range(Fleet_Size):
        if len(Buses[b]['Dest']) == 0:
            continue
        else:
            busAlight = copy.deepcopy(Buses[b]['Alight'])
            
            for person , time in busAlight.items():
                passenger.append(person)
                Bus_Selected.append(b)
                Origin_List.append(Buses[b]['Origin'][person])
                Dest_List.append(Buses[b]['Dest'][person])
                
                EntryTime.append(Buses[b]['Entry'][person])
                BoardingTime.append(Buses[b]['Board'][person])
                AlightingTime.append(Buses[b]['Alight'][person])
                waitingTime.append(timedelta2time(BoardingTime[-1] - EntryTime[-1]))
                TravelTime.append(timedelta2time(AlightingTime[-1] - BoardingTime[-1]))
                
                Buses_Route_Person.append(Buses[b]['Routes']['Person'])
                Buses_Route_Loc.append(Buses[b]['Routes']['Loc'])
                
                try:
                    Buses[b]['Entry'].pop(person)
                    Buses[b]['Board'].pop(person)
                    Buses[b]['Alight'].pop(person)
                    Buses[b]['Dest'].pop(person)
                    Buses[b]['Origin'].pop(person)
                    Buses[b]['Promise_Alight'].pop(person)
                    Buses[b]['Promise_Board'].pop(person)
                    

                except KeyError:
                    print('Key Not In The Dictionary')

print(len(passenger) , len(EntryTime) , len(waitingTime) , len(BoardingTime) , len(TravelTime) , len(AlightingTime) , len(Bus_Selected), len(Buses_Route_Person), len(Buses_Route_Loc))
```
